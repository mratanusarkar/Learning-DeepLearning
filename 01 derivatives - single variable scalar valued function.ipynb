```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Derivatives: scalar valued function with one variable f(x)

```python
# define a scalar valued function f(x)
def f(x):
    return 3 * x**2 - 4*x + 5
```

```python
# we can pass x to f(x) and get y
f(1.0)
```

```python
x_values = np.arange(-5, 5, 0.25)
y_values = f(x_values)
plt.plot(x_values, y_values)
```

```python
# for f(x) = 3 * x**2 - 4*x + 5
# using chain rule and computing manually
# f'(x) will be:
#
# = 3 * 2 * x - 4 + 0
# = 6 * x - 4

def grad_f_chain(x):
    return 6 * x - 4
```

```python
# we can pass x to f'(x) and get derivative at x
grad_f_chain(1.0)
```

```python
x_values = np.arange(-5, 5, 0.25)
y_values = grad_f_chain(x_values)
plt.plot(x_values, y_values)
```

```python
# let's try deriving f'(x) from limits and first principles
#
# L = lim h->0 [f(x + h) - f(x)] / h
# provided the fn is differentiable at x, and is continuous... as h tends to 0
# f is differentiable at x, i.e the limit L exists, then L is called derivative of f at x
# hence, f'(x) = [f(x + h) - f(x)] / h
```

```python
# derivative using first principles
def grad_f_lim(x, h=0.001):
    return (f(x + h) - f(x)) / h
```

```python
# let's compare the results
x = 3.0
print("f'(x) using first principles:", grad_f_lim(x, 0.000001))
print("f'(x) using chain rule      :", grad_f_chain(x))
```

```python
x_values = np.arange(-5, 5, 0.25)
y_values = grad_f_lim(x_values, 0.0001)
plt.plot(x_values, y_values)
```

```python
x_values = np.arange(-5, 5, 0.25)
plt.plot(x_values, f(x_values), 'b', x_values, grad_f_chain(x_values), 'g', x_values, grad_f_lim(x_values), 'r+')
```

```python
# as we see the derivatives by both the methods are almost similar,
# since no one sit and derive the grads manually, and for neural networks, it would be a massive equation
# we go with the first principles to calculate the derivatives or the grads
```
